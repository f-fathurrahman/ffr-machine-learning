# Gaussian density function

In 1d:
$$
p(x|\mu,\sigma^2) = \mathcal{N}(\mu,\sigma^2) = \frac{1}{\sigma\sqrt{2\pi}}
\exp\left\{-\frac{1}{2\sigma^2}(x-\mu)^2\right\}
$$

- $\mu$: mean
- $\sigma^2$: variance

Multivariate Gaussian:
$$
p(\mathbf{x}) =
\frac{1}{(2\pi)^{D/2}\left|\Sigma\right|^2}
\exp\left\{
-\frac{1}{2}\left(\mathbf{x} - \boldsymbol{\mu}\right)^{\mathrm{T}}
\Sigma^{-1}
\left(\mathbf{x} - \boldsymbol{\mu}\right)
\right\}
$$

- $\mathbf{x} = \left[ x_1, x_2, \ldots, x_{D} \right]$: $D$-dimensional vector
- $\Sigma$: $D \times D$ covariance matrix

# Beta density function

$$
p(r) = \frac{\Gamma(\alpha + \beta)}{\Gamma(a)\Gamma(\beta)}
r^{\alpha-1}(1-r)^{\beta-1}
$$

```julia
using SpecialFunctions
using PGFPlotsX
```

```julia
function beta_pdf(r, alpha, beta)
  return gamma(alpha+beta)/gamma(alpha)/gamma(beta) * r^alpha * (1-r)^beta
end
```

```julia
x = 0:0.01:1
a = 1.0
b = 1.0
y = beta_pdf.(x, 1.0, 0.5)
@pgf Axis(
    PlotInc(Coordinates(x,y)),
)
```
